```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from statsmodels.distributions.empirical_distribution import ECDF
from scipy.stats import gilbrat
%matplotlib inline
```

```python
# зададим логнормальную случайную величину, s=1 (Gilbrat random variable)
gilb_rv = sts.gilbrat(loc=0, scale=1)
# сгенерируем 1000 значений
sample = gilb_rv.rvs(size=1000)
# генерация значений аргумента
x = np.linspace(0,10,1000)
```

```python
# задаём плотность вероятности как функцию от вектора значений аргументов
pdf = gilb_rv.pdf(x)
# строим её график
plt.plot(x, pdf)
# подписываем оси
plt.ylabel('f(x)')
plt.xlabel('x')
# строим гистограмму
a = plt.hist(sample, bins=15, density=True)
```

```python
# вычысляем матожидание и дисперсию
mean, var = gilbrat.stats(loc=0, scale=1, moments='mv')
print(mean, var)
```

```python
# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mean, scale=(var/10)**(1/2))

# задаём плотность вероятности как функцию от вектора значений аргументов
x = np.linspace(-2,6,1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

# подписываем оси
plt.ylabel('f(x)')
plt.xlabel('x')

# приготовим список из 1000 выборок
sample_mean = [0 for i in range(1000)]
# сгенерируем 1000 выборок длины 10
for i in range(1000):
    sample = gilb_rv.rvs(size=10)
    # Посчитаем выборочное среднее для каждой выборки и запишем в список
    for item in sample:
        sample_mean[i] += item
    sample_mean[i]/=10
    
# строим гистограмму
a = plt.hist(sample_mean, bins=20, density=True)


```
