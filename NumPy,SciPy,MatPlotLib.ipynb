# Все вместе

```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
from scipy import interpolate
```

```python
x = np.arange(0, 10, 2)
y = np.exp(-x / 3.0) + np.random.randn(len(x)) * 0.05
print x
print y
```

```python
f = interpolate.interp1d(x, y, kind = 'quadratic')
xnew = np.arange(0, 8, 0.1)
ynew = f(xnew)
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```
