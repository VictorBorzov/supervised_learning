```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh
from scipy.stats import norm
%matplotlib inline
```

## Общее описание распределения Рэлея

**Распределение Рэлея** — это распределение вероятностей случайной величины $X$ с плотностью

$f(x;\sigma )={\frac {x}{\sigma ^{2}}}\exp \left(-{\frac {x^{2}}{2\sigma ^{2}}}\right),x\geqslant 0,\sigma >0,$

где $\sigma$  — параметр масштаба. Соответствующая функция распределения имеет вид

$\mathsf P(X\leqslant x)=\int \limits _{0}^{x}f(\xi )\,d\xi =1-\exp \left(-{\frac {x^{2}}{2\sigma ^{2}}}\right),x\geqslant 0.$


Введено впервые в 1880 г. Джоном Уильямом Стреттом (лордом Рэлеем) в связи с задачей сложения гармонических колебаний со случайными фазами.

### Применение
 - В задачах о пристрелке пушек. Если отклонения от цели для двух взаимно перпендикулярных направлений нормально распределены и некоррелированы, координаты цели совпадают с началом координат, то, обозначив разброс по осям как $X$ и $Y$, получим выражение для величины промаха в виде $R={\sqrt {{{X}^{2}}+{{Y}^{2}}}}$. В этом случае величина $R$ имеет распределение Рэлея.
 - В радиотехнике для описания амплитудных флуктуаций радиосигнала.
 - Плотность распределения излучения абсолютно чёрного тела по частотам.

###  Распределения Рэлея по одной выборке из 500 событий

```python
n = 500
rv = rayleigh.rvs(size=n)
```

#### Построение гистограммы и графика плотности распределения

```python
plt.figure(figsize=(13,10))
x = np.arange(0, 5, 0.01)

plt.hist(rv, normed=True, bins=20)
plt.plot(x, rayleigh.pdf(x))
plt.show()
```

### Применение ЦПТ
#### Генерирование 1000 выборок по *count* событий в каждой выборке 

```python
count = 5
rv = rayleigh.rvs(size=(1000,count))
```

#### Построение диаграммы распределения средних

```python
means = rv.mean(axis=1)
plt.figure(figsize=(13,10))
plt.hist(means, normed=True, bins=30)
plt.show()
```

#### Среднее (Мат.ожидание) 
$E = \sigma\sqrt{\pi/2}$

```python
E = 1*((np.pi)/2)**0.5
E
```

#### Дисперсия
$D = \frac{4 - \pi }{2}\sigma^2$

```python
D = ((4 - np.pi)/2)*1**2
D
```

#### Гистограмма выборочных средних для выбранного распределения и построение графика плотности вероятности для нормального распределения

```python
E_norm = E
var_norm = D/count
rv_norm = norm(E_norm, var_norm**0.5)
```

```python
plt.figure(figsize=(13,10))
x = np.arange(0.2, 2.3, 0.001)
plt.hist(means, normed = True, bins=30)
plt.plot(x, rv_norm.pdf(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Rayleigh means for 1000 * 5 & Normal Distribution pdf')
plt.show()
```

### Кол-во событий в выборке = 10
