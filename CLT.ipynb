```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Экспоненциальное распределение

$ f_{\xi}(x)=e^{-x}I_{x>0} $

$F_{\xi}(x)=(1-e^{-x})I_{x>0}$

Переменная dat - выборка

```python
dat = sts.expon.rvs(size=1000)
```

### Эмпирическое и теоретическое распределения

```python
xlab = np.linspace(0, 10, 20)
plt.hist(dat, normed=True, color='Orange', bins=20)
plt.plot(xlab, sts.expon.pdf(xlab))
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

## ЦПТ

$ \sqrt{n}\frac{\bar{X}-\mu}{\sigma} \to\ Z \sim N(0, 1)$

$ \bar{X} \to Q \sim N(\mu, \frac{\sigma^2}{n})$

$ X_i \sim exp(1) \Rightarrow \mu=1, \frac{\sigma^2}{n}=\frac{1}{n}$

$\xi \sim N(\mu, \sigma^2) \Rightarrow f_{\xi}(x)=\frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

```python
f = lambda x: 1./np.sqrt(2 * np.pi * (1./n)) * np.exp(-(x - 1)**2/(2. * (1./n)))
```

```python
N = [4, 25, 49]
for n in N:
    T = [None] * 1000
    T = list(map(lambda x: np.mean(sts.expon.rvs(size=n)), T))
    print 'n = %d' % n
    E = 1
    Var = 1./n
    x = np.linspace(np.min(T), np.max(T), 1000)
    
    plt.hist(T, normed=True, bins=20, color='Orange')
    plt.plot(x, list(map(f, x)))
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.show()
```

Среднее стремится к нормальному распределению со скоростью $\sqrt{n}$. На графиках изображены графики средних из $n=4, 25, 49$ измерений. Поэтому можно сказать, что третье измерение в $ \frac{7}{5}=1.4$ лучше второго и в $\frac{7}{2}=3.5$ лучше первого.
